### Pre-selection of Variables

**Variables to DROP:**
- Height and Weight: Redundant information since BMI is already calculated
- Total Body Water, ECW, ICW, ECF/TBW: Multiple body water measurements provide redundant information
- Bone Mass, Muscle Mass, Lean Mass, Body Protein Content: Body composition details not directly related to gallstone formation
- AST, ALT, ALP, Creatinine, GFR, CRP, Hemoglobin: Secondary clinical markers; focusing on primary metabolic risk factors
- Comorbidity, CAD, Hypothyroidism, Hyperlipidemia: Multiple disease indicators; keeping only Diabetes as key metabolic disease
- Vitamin D: Limited direct connection to gallstone pathophysiology
- VFR, Obesity %: Redundant with Visceral Fat Area and BMI

**Variables to KEEP:**
- Gallstone Status: Response variable
- Age, Gender: Key demographic predictors
- BMI: Primary obesity measure
- Glucose, Diabetes: Glucose metabolism indicators
- Total Cholesterol, LDL, HDL, Triglycerides: Complete lipid profile (cholesterol forms gallstones)
- Visceral Fat Area: Direct measure of abdominal obesity
- Hepatic Fat Accumulation: Liver fat related to metabolic syndrome

**Rationale:** We focus on variables directly related to metabolic syndrome and gallstone pathophysiology while removing redundant measurements.

## Section 3: Exploratory Data Analysis and Visualization

```r
gallstone <- gallstone_raw %>%
  select(
    gallstone = `Gallstone Status`,
    age = Age,
    gender = Gender,
    bmi = `Body Mass Index (BMI)`,
    glucose = Glucose,
    diabetes = `Diabetes Mellitus (DM)`,
    cholesterol = `Total Cholesterol (TC)`,
    ldl = `Low Density Lipoprotein (LDL)`,
    hdl = `High Density Lipoprotein (HDL)`,
    triglycerides = `Triglyceride`,
    visceral_fat = `Visceral Fat Area (VFA)`,
    hepatic_fat = `Hepatic Fat Accumulation (HFA)`
  ) %>%
  mutate(
    gallstone = as.numeric(gallstone),
    age = as.numeric(age),
    gender = as.factor(gender),
    bmi = as.numeric(bmi),
    glucose = as.numeric(glucose),
    diabetes = as.factor(diabetes),
    cholesterol = as.numeric(cholesterol),
    ldl = as.numeric(ldl),
    hdl = as.numeric(hdl),
    triglycerides = as.numeric(triglycerides),
    visceral_fat = as.numeric(visceral_fat),
    hepatic_fat = as.factor(hepatic_fat)
  ) %>%
  drop_na()

dim(gallstone)
head(gallstone)
```

```r
# Set plot size
options(repr.plot.width = 12, repr.plot.height = 8)

# Create visualization
exploratory_plot <- gallstone %>%
  filter(hepatic_fat %in% c("0", "1", "2")) %>%
  mutate(
    gallstone_status = factor(gallstone, labels = c("No Gallstone", "Gallstone")),
    diabetes_status = factor(diabetes, labels = c("No Diabetes", "Diabetes")),
    hfa_level = factor(hepatic_fat, 
                       levels = c("0", "1", "2"),
                       labels = c("None", "Mild", "Moderate"))
  ) %>%
  ggplot(aes(x = age, y = bmi, color = triglycerides, shape = gallstone_status)) +
  geom_point(size = 3, alpha = 0.7) +
  scale_color_gradient(low = "#3B9AB2", high = "#E1AF00", name = "Triglycerides\n(mg/dL)") +
  scale_shape_manual(values = c(16, 17), name = "Gallstone\nStatus") +
  facet_grid(diabetes_status ~ hfa_level) +
  labs(
    title = "Metabolic Risk Factors for Gallstone Disease",
    subtitle = "Exploring Age, BMI, Triglycerides, Diabetes, and Hepatic Fat Accumulation",
    x = "Age (years)",
    y = "Body Mass Index (kg/m²)"
  ) +
  theme_minimal() +
  theme(
    text = element_text(size = 12),
    plot.title = element_text(face = "bold", size = 14),
    strip.text = element_text(face = "bold")
  )

exploratory_plot
```

**Variable Selection for Visualization:**

From the 12 variables in our cleaned dataset, I selected 5 variables for this visualization: age, BMI, triglycerides, diabetes status, and hepatic fat accumulation. I chose these because they represent different aspects of metabolic health body composition (BMI), blood lipids (triglycerides), blood sugar regulation (diabetes), and liver fat (hepatic fat) while keeping the plot interpretable. I filtered hepatic fat to only show levels 0, 1, and 2 because levels 3 and 4 had very few observations (27 total out of 319), which would create mostly empty panels.

### Visualization Interpretation

**Why this plot is relevant:**

This visualization explores five variables at once by using position (age, BMI), color (triglycerides), shape (gallstone status), and facets (diabetes, hepatic fat). Faceting by diabetes and hepatic fat lets me see whether gallstone patterns differ across these metabolic conditions. This multi-variable view is useful because gallstone disease likely involves multiple risk factors rather than just one.


**Results obtained:**

The plot shows that gallstone patients (triangles) and non-gallstone patients (circles) overlap considerably in age and BMI across all six panels. Most patients have low-to-moderate triglyceride levels (blue/teal colors), with only a few showing high levels (yellow colors). The top row (No Diabetes) has far more patients than the bottom row (Diabetes), and the Moderate hepatic fat panel appears to have the most observations overall. Among diabetes patients, gallstone cases seem to cluster more in the Moderate hepatic fat panel at higher BMI values. Overall, no single variable clearly separates gallstone from non-gallstone patients, suggesting that multiple factors may be involved.


**What I learned:**

Three things stand out from this visualization. First, the overlap between gallstone and non-gallstone patients across age and BMI suggests that these factors alone won't predict gallstone disease well. Second, the Moderate hepatic fat panel contains the most data and shows a noticeable concentration of gallstone cases, especially among diabetes patients. Third, most patients have relatively low triglyceride levels regardless of gallstone status, so extremely high triglycerides don't seem necessary for gallstone formation in this dataset. These patterns suggest I'll need a multivariable model to understand how these factors work together.
